```python
import math
print("factorial of 5 is ",math.factorial(5))
print("pi : ",math.pi)
print("floor of 5.78634 is ",math.floor(5.78634))
print("ceil of 5.78634 is ", math.ceil(5.78634))    
print("sin(1) is ",math.sin(1))
print("natural log of 10 is ", math.log(10))        # natural log (base e)
print("log base 10 of 10 is ", math.log10(10))      # log base 10
print("log base 2 of 8 is ", math.log(8, 2))        # log base 2
print("square root of 49 is ",math.sqrt(49)) 
print("tan(45 degrees) is ", math.tan(math.radians(45))) # converting degrees to radians
print("5 to the power of 3 is ",math.pow(5,3))
```
