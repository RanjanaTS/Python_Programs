```python
# Example of composing expressions
import math
x=5
result=math.sqrt(x+4*2) # The argument itself is an expression
print(result)
```

```python
# Example of composition function 
def area(s):
    return s*s
def volume(s):
    return area(s)*s
s=2
print("area of square of side 2 : ",area(s))
print("volume of square of side 2 : ",volume(s))
```

```python
# Example of composition function 
import math
def square(x):
    return x*x
def root(y):
    return math.sqrt(y)
# Composition
print(root(square(5)))   # sqrt(25)=5
```

```python
# Example of composition function 
def trim(text):
    return text.strip()
def to_upper(text):
    return text.upper()
# Composition
print(to_upper(trim("   hello world   ")))  # "HELLO WORLD"
```
